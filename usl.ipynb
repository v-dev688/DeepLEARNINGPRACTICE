```python
# outlier removal
def remove_ouliers_iqr(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df=df[~((df<(Q1-1.5*IQR))|(df>Q3 + (1.5*IQR))).any(axis=1)]
    return df
```

```python
df=remove_ouliers_iqr(df)
```

```python
#a. How will you decide when to apply PCA based on the correlation? (2 marks)

#PCA is applied for following cases

# 1. High dimensional data
# 2. Multicollinarity excisting
# 3. Noise Reduction
# 4. Visualizations - Projecting High dimensional data
# 5. Computational Efficiency
# 6. Model Performance Improvement
```

```python
df.columns
```

```python
from sklearn.decomposition import PCA

X=df
pca=PCA()
X_pca=pca.fit_transform(X)
```

```python
import numpy as np
cum_var_ratio=np.cumsum(pca.explained_variance_ratio_)
```

```python
num_components=np.argmax(cum_var_ratio>0.9)+1
num_components
```

```python
columns_pca=[f"PC{i+1}" for i in range(X_pca.shape[1])]
df_pca=pd.DataFrame(X_pca, columns=columns_pca)
```

```python
#d. Visualize the spread of data across PCA components. (2 marks)
import seaborn as sns
import matplotlib.pyplot as plt

fig,ax= plt.subplots(figsize=(20,12))
sns.heatmap(df_pca.corr(),annot= True, cmap="coolwarm",fmt=".2f",ax=ax)
plt.show()
```

```python
#multicollinarity  is reduced but only excists between two components
```

```python
#e. Check for outliers in the PCA data and treat the same. (2 marks)

plt.figure(figsize=(12,6))
sns.boxplot(df_pca.corr())
plt.show()
```

```python
#a. Apply Clustering and visualize the spread of data (15 marks)

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_std=scaler.fit_transform(X)

kmeans= KMeans(n_clusters=5, random_state=42)
kmeans_labels=kmeans.fit_predict(X_std)

agg_clusuring =AgglomerativeClustering(n_clusters=5)
agg_labels=agg_clusuring.fit_predict(X_std)
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.scatterplot(x=X_std[:,0],y=X_std[:,1],hue=kmeans_labels, palette="viridis",legend="full")
plt.title("K_ Means")
plt.show()
```

```python
#b. Find the optimal number of cluster Value. (8 marks)

# For finding this optimal values i am using Elbow plots

inertia_values=[]
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(X_std)
    inertia_values.append(kmeans.inertia_)
    
plt.plot(range(1,20),inertia_values,marker="o")
plt.xlabel("Numer of clusters")
plt.ylabel("Inertia values")
plt.show()

#between 17 to 18 
```
