```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
img = load_img('/content/drive/MyDrive/rp.jpg')  # this is a PIL image
x = img_to_array(img)  # this is a Numpy array with shape (3, 150, 150)
x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 3, 150, 150)
```

```python
i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='/content/drive/MyDrive/DATA_AUGMENTATION', save_prefix='pattinson', save_format='jpeg'):
    i += 1
    if i > 20:
        break
```

# Sample output image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

def display_images(image_folder):
    # Get a list of all image files in the specified folder
    image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

    # Plot each image in a 5x4 grid
    rows, cols = 5, 4
    plt.figure(figsize=(15, 12))  # Adjust the figure size as needed
    for i, image_file in enumerate(image_files, start=1):
        img_path = os.path.join(image_folder, image_file)
        img = mpimg.imread(img_path)

        plt.subplot(rows, cols, i)
        plt.imshow(img)
        plt.title(f"Image {i}")
        plt.axis("off")

    plt.tight_layout()
    plt.show()
image_folder_path = "/content/drive/MyDrive/DATA_AUGMENTATION"

# Display the images in a 5x4 grid
display_images(image_folder_path)
```
